```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from fspb.config import BLD_APPLICATION
from fspb.bands.linear_model import ConcurrentLinearModel
```

```python
x_train = pd.read_pickle(BLD_APPLICATION / "x_train.pickle")
x_pred = pd.read_pickle(BLD_APPLICATION / "x_pred.pickle")

y_train = pd.read_pickle(BLD_APPLICATION / "y_train.pickle")
y_pred = pd.read_pickle(BLD_APPLICATION / "y_pred.pickle")
```

```python
model = ConcurrentLinearModel()
model.fit(x_train, y_train)
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
time_grid = np.linspace(0, 1, x_train.shape[2])
ax.plot(time_grid, y_pred.T, label="True", color="blue")
ax.plot(time_grid, model.predict(x_pred).T, label="Predicted", color="red")
```
